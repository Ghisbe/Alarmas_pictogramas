```python
url = "https://drive.google.com/uc?id=1EoEqJHbGgVSDRIg5Fv_m_OCGoBJOLFkp"
```

```python
output = 'alertas_clima.csv'
gdown.download(url, output, quiet=False)

df = pd.read_csv(output)
```

```python
df
```

```python
# cantidad de valores nulos en cada columna
nulos_por_columna = df.isnull().sum()

# cantidad de valores nulos en todo el DataFrame
total_nulos = df.isnull().sum().sum()

print("Valores nulos por columna:")
print(nulos_por_columna)
print("\nTotal de valores nulos en el DataFrame:", total_nulos)
```

```python
# elimina registros con valores nulos en todo el DataFrame
df_sin_nulos = df.dropna(subset=['description', 'instruction'])
```

```python
df_sin_nulos.info()
```

```python
df_sin_nulos
```

```python
# cantidad de registros con instrucción  \n\n
filas_nn = df_sin_nulos[df_sin_nulos['instruction'] == '\n\n']
cantidad_nn = len(filas_nn)
cantidad_nn
```

```python
# elimina registros con valores \n\n en el campo instrucción
df_filtrado = df_sin_nulos[df_sin_nulos['instruction'] != '\n\n']
```

```python
df_filtrado.info()
```

```python
df_filtrado
```

```python
#revisión de valores únicos
valores_unicos_urgency = df_filtrado['urgency'].unique()
valores_unicos_urgency
```

```python
# Elimina los registros donde el campo 'urgency' tiene el valor 'Unknown'
df_filtrado_sin_unknown = df_filtrado.loc[df_filtrado['urgency'] != 'Unknown']
```
