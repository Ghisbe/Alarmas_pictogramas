```python
output = 'alertas_procesado.csv'
gdown.download(url, output, quiet=False)

# Leer el archivo CSV en un DataFrame de pandas
df_alertas = pd.read_csv(output)
```

Se elige aquellas alertas que contienen la palabra storm.

```python
# Filtrar las filas que contienen la palabra 'storm' en la columna 'event'
filtro_df = df_alertas[df_alertas['event'].str.contains('storm', case=False, na=False)]
```

```python
# Verificar las columnas del DataFrame original
print("Columnas del DataFrame original:")
print(df_alertas.columns)

# Verificar las columnas del DataFrame resultante después del filtro
print("\nColumnas del DataFrame resultante después del filtro:")
print(filtro_df.columns)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
# Definir la función para lematizar el texto con spaCy
def lemmatize_text_spacy(text):
    # Lematizar el texto utilizando spaCy
    doc = nlp(text)
    lemmatized_text = " ".join([token.lemma_ for token in doc])
    return lemmatized_text
```

```python
# Aplicar la lematización por lotes a la columna 'description_no_stopwords'
batch_size = 1000
num_batches = len(filtro_df) // batch_size + 1

for i in range(num_batches):
    batch_start = i * batch_size
    batch_end = min((i + 1) * batch_size, len(filtro_df))
    filtro_df.loc[batch_start:batch_end, 'lemmatized_description'] = filtro_df.loc[batch_start:batch_end, 'description_no_stopwords'].apply(lemmatize_text_spacy)
```

```python
filtro_df['lemmatized_description']
```

```python
nan_sum = filtro_df['lemmatized_description'].isna().sum()

# Imprimir la suma de valores NaN
print("Número de valores NaN en la columna 'description_lemmatized':", nan_sum)
```

Se eliminan los valores nan

```python
# Eliminar filas con NaN en la columna 'description_lemmatized'
filtro_df = filtro_df.dropna(subset=['lemmatized_description'])
```
