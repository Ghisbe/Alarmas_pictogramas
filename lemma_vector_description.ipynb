```python
from nltk.stem import WordNetLemmatizer
```

```python
# Inicializar el lematizador
lemmatizer = WordNetLemmatizer()

# Función para lematizar el texto en lotes
def lemmatize_text_batch(texts):
    lemmatized_texts = []
    for text in texts:
        words = text.split()
        lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
        lemmatized_texts.append(' '.join(lemmatized_words))
    return lemmatized_texts
```

```python
# Aplicar la lematización por lotes a la columna 'description_no_stopwords'
batch_size = 1000  # Tamaño del lote
num_batches = len(df_alertas) // batch_size + 1  # Calcular el número de lotes

for i in range(num_batches):
    batch_start = i * batch_size
    batch_end = min((i + 1) * batch_size, len(df_alertas))  # Calcular el índice de fin del lote
    df_alertas.loc[batch_start:batch_end, 'lemmatized_description'] = lemmatize_text_batch(
        df_alertas.loc[batch_start:batch_end, 'description_no_stopwords']
    )
```
