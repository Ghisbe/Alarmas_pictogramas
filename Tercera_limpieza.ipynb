```python
from google.colab import drive
drive.mount('/content/drive')
import zipfile
import pandas as pd
```

```python
ruta = '/content/drive/MyDrive/Procesamiento de audio y habla/alertas_clima.zip'

# Nombre del archivo CSV dentro del archivo ZIP
nombre_archivo_csv = 'alertas_clima.csv'

with zipfile.ZipFile(ruta, 'r') as zip_ref:
    with zip_ref.open(nombre_archivo_csv) as csv_file:
        df_alertas = pd.read_csv(csv_file)
```
