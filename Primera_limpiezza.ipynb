Agrupacion de las categorias por tipo de alertas

```python
palabras_clave = {
    'MeteorolÃ³gica': ['Weather', 'Flood', 'Hurricane', 'Tornado', 'Blizzard', 'Snow', 'Wind', 'Heat', 'Cold', 'Freeze', 'Storm', 'Tsunami','Cyclone','Fog','Avalanche'],
    'Civil': ['Civil', 'Law Enforcement', 'Emergency', 'Shelter in Place', 'Evacuation', 'Amber Alert', 'Missing', 'Curfew','Wanted','Radiological Hazard','Elevator Fire','Silver Alert',
              'chemical spill'],
}

def asignar_tipo(alerta):
    for tipo, palabras in palabras_clave.items():
        if any(palabra.lower() in alerta.lower() for palabra in palabras):
            return tipo
    return 'Otra'  # Si ninguna palabra clave coincide

nuevo_df['tipo_alerta'] = nuevo_df['event'].apply(asignar_tipo)

alertas_por_tipo = nuevo_df['tipo_alerta'].value_counts()

print(alertas_por_tipo)
```

```python
alertas_meteorologicas = nuevo_df[nuevo_df['tipo_alerta'] == 'MeteorolÃ³gica']
alertas_unicas_meteorologicas = alertas_meteorologicas['event'].unique()
print(alertas_unicas_meteorologicas)

cantidad_alertas_meteorologicas = len(alertas_unicas_meteorologicas)
print("Cantidad de alertas meteorolÃ³gicas Ãºnicas:", cantidad_alertas_meteorologicas)
```

```python
alertas_civiles = nuevo_df[nuevo_df['tipo_alerta'] == 'Civil']
alertas_unicas_civiles = alertas_civiles['event'].unique()
print(alertas_unicas_civiles)

cantidad_alertas_civiles = len(alertas_unicas_civiles)
print("Cantidad de alertas civiles Ãºnicas:", cantidad_alertas_civiles)
```

```python
alertas_otra = nuevo_df[nuevo_df['tipo_alerta'] == 'Otra']
alertas_unicas_otra = alertas_otra['event'].unique()
print(alertas_unicas_otra)

cantidad_alertas_otras = len(alertas_unicas_otra)
print("Cantidad de otras alertas:", cantidad_alertas_otras)
```
