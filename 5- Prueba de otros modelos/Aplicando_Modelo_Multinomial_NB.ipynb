### Modelo Multinomial NB
Este modelo de clasificacion probabilístico basado en el Teorema de Bayes, utilizado principalmente para problemas de clasificación de texto y análisis de datos discretos. Este modelo es una variante del clasificador de Naive Bayes que se adapta particularmente bien a datos que representan frecuencias de ocurrencia de eventos, como el conteo de palabras en documentos. En este caso se tiene la descripcion del evento (variable x) y se busca predecir el tipo de evento en el que se clasifica (variable y).

```python
from sklearn.model_selection import train_test_split

x = df_filtrado['description']
y = df_filtrado['event']

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2)

# Verificar dimensiones después de la división
print(f"Dimensiones de X_train: {X_train.shape}")
print(f"Dimensiones de Y_train: {Y_train.shape}")
print(f"Dimensiones de X_test: {X_test.shape}")
print(f"Dimensiones de Y_test: {Y_test.shape}")
```

Para aplicar este modelo, se vectoriza primero el texto de la variable X a traves del metodo CountVectorizer()

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

X_train_transformed = vectorizer.fit_transform(X_train)
X_test_transformed = vectorizer.transform(X_test)
```

Asi se importa el modelo y se entrena:

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

model = MultinomialNB()
model.fit(X_train_transformed, Y_train)
```

Se busca establecer la prediccion del modelo

```python
Y_predict = model.predict(X_test_transformed)
print(f'La probalidad del modelo es de {round((metrics.accuracy_score(Y_test, Y_predict))*100,2)}%')
```

Se observa que se tiene una alta probabilidad de acierto este modelo de clasificacion.

Se genera un reporte sobre la clasificacion:

```python
from sklearn.metrics import confusion_matrix, classification_report

# Generar un reporte de clasificación
report = classification_report(Y_test, Y_predict)
print(report)
```

Se genera un dataframe con 10 ejemplos respecto de las predicciones y su acierto o error.

```python
import pandas as pd

def mostrar_resultados_dataframe(X_test, Y_test, Y_predict, n_ejemplos=10):
    resultados = []

    for i in range(n_ejemplos):
        fila = {
            'Ejemplo': X_test.index[i],
            'Etiqueta Real': Y_test.iloc[i],
            'Predicción': Y_predict[i],
            'Correcto': Y_test.iloc[i] == Y_predict[i]
        }
        resultados.append(fila)

    df_resultados = pd.DataFrame(resultados)
    return df_resultados

# Mostrar los resultados en un DataFrame
df_resultados = mostrar_resultados_dataframe(X_test, Y_test, Y_predict, n_ejemplos=10)
df_resultados
```
